```python
import numpy as np
import matplotlib.pyplot as plt

#Configuraion general de los plots revisar documentacion
plt.rcParams['figure.figsize'] = (4,3)
plt.rcParams['figure.dpi'] = 150
```

```python
# Basico
xvals = np.arange(0,10,0.1)
plt.plot(xvals)
```

```python
#Ejes
plt.plot(xvals,np.sin(xvals))
plt.plot(xvals,np.cos(xvals))
```

```python
plt.xlabel("$x$")
plt.ylabel("$y$")
plt.plot(xvals,np.tan(xvals))
```

```python
plt.plot(xvals,np.sin(xvals),label=r"$y =sin(x)$",color='r')
plt.plot(xvals,np.cos(xvals),label=r"$y =cos(x)$",color='b')
plt.plot(xvals,np.tan(xvals),label=r"$y =tan(x)$",color='g')

plt.xlabel("$x$")
plt.ylabel("$y$")
plt.legend()
```

```python
plt.plot(xvals,np.sin(xvals),label=r"$y =sin(x)$",color='k')
plt.plot(xvals,np.cos(xvals),label=r"$y =cos(x)$",color='b',marker='x')
plt.xlabel("$x$")
plt.ylabel("$y$")
plt.legend(loc='upper left')
```

```python
plt.plot(xvals,np.sin(xvals),'g--',label=r"$y =sin(x)$")
plt.plot(xvals,np.cos(xvals),'bx',label=r"$y =cos(x)$",color='b',marker='o')
plt.xlabel("$x$")
plt.ylabel("$y$")
plt.legend(loc='upper left')
```

```python
plt.plot(xvals,np.sin(xvals),'g--',label=r"$y =sin(x)$")
plt.plot(xvals,np.cos(xvals),'bx',label=r"$y =cos(x)$",color='b',marker='o')
plt.xlim(0,2)
plt.xlim(0,1)
plt.xlabel("$x$")
plt.ylabel("$y$")
plt.legend(loc='upper left')
```
