```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
plt.rcParams['figure.figsize'] = (4,3)
plt.rcParams['figure.dpi'] = 150
```

```python
xvals = np.arange(0,10,0.1)
plt.plot(xvals,np.sin(xvals))
plt.xlabel(r'$x$')
plt.ylabel(r'$sin(x)$')
```

```python
idx = 0 
for marker in ('-','--',':','-.'):
    plt.plot(xvals,np.sin(xvals)+idx,''.join(('k',marker)))
    idx += 1
```

```python
xvals = np.arange(0,10,0.1)
plt.plot(xvals,np.sin(xvals),'k.--',markevery=40)
plt.xlabel(r'$x$')
plt.ylabel(r'$sin(x)$')
```

```python
xvals = np.arange(0,10,0.1)
plt.plot(xvals,np.sin(xvals),'k.--',markersize=5,markevery=3)
plt.xlabel(r'$x$')
plt.ylabel(r'$sin(x)$')
```
