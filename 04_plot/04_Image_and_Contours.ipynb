```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
plt.rcParams['figure.figsize'] = (4,3)
plt.rcParams['figure.dpi'] = 150
```

```python
from scipy.ndimage.filters import  gaussian_filter
rands2d = gaussian_filter(np.random.normal(size=(512,512)),sigma=10)
type(rands2d)
```

```python
plt.imshow(rands2d)
plt.colorbar()
```

```python
plt.imshow(rands2d,extent=[-200,100,-100,100])
plt.colorbar()
```

```python
plt.imshow(rands2d,vmax=0.3,vmin=-0.05) #Limitar valores minimos
plt.colorbar()
```

```python
plt.imshow(rands2d,aspect=0.4) #Menos o mas horizontal
```

```python
plt.imshow(rands2d,aspect=1.4) #Menos o mas horizontal
```
