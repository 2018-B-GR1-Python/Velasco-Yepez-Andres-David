```python
import numpy as np
from scipy import misc
import matplotlib.pyplot as plt
```

## Generar un rompecabezas

```python
mapache = misc.face()  # Cara de un mapache
def get_puzzle_imagen(numero_partes):
    mapache_partes = np.vsplit(mapache,numero_partes)
    piezas_indices = np.random.permutation(numero_partes)
    mapache_mezclado = np.concatenate(list(map(lambda x: mapache_partes[x],piezas_indices)))
    plt.imshow(mapache_mezclado)
    plt.show()
```

```python
get_puzzle_imagen(8)
```

## Distorcionar imagen

```python
mapache = misc.face()
def distorcionar_imagen(key,imagen):
    return np.array(list(map(lambda x:(x*key)%255,imagen)))
plt.imshow(distorcionar_imagen(4,mapache))
plt.show()
```

## Cifrar palabras

```python
def cifrado_cesar(key,cadena):
    abecedario_ascii = np.arange(65,91)
    abecedario = list(map(lambda x: chr(x),abecedario_ascii))
    abecedario_desplazado = np.array(list(map(lambda x: x+key if (x+key) <90 else (x+key)%91,abecedario_ascii)),dtype=int)
    map_abc =np.stack((abecedario,abecedario_desplazado),axis=1)
    diccionario =  {k : v for k,v in map_abc}
    return ''.join( chr(int(diccionario[a.upper()])) for a in cadena )
```

```python
cadena='Andres'
print(cifrado_cesar(4,cadena))
```
