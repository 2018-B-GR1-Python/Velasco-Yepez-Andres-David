```python
import numpy as np
from scipy import misc
```

```python
mapache = misc.face()  # Cara de un mapache
```

```python
mapache.shape
```

```python
type(mapache)
```

```python
import matplotlib.pyplot as plt
```

```python
plt.imshow(mapache)
print(mapache[0][0][1])
print(mapache[0][0][1])
print(mapache[0][0][2])
plt.show()
```

```python
mapache_parcial = mapache[385:,512:,:]
```

```python
plt.imshow(mapache_parcial)
plt.show()
```

```python
mapache_parte_uno,mapache_parte_dos =np.split(mapache,2,axis=1)
plt.imshow(mapache_parte_uno)
plt.show()
```

```python
mapache_frankenstein = np.concatenate((mapache_parte_uno,mapache_parte_dos),axis=1)
plt.imshow(mapache_frankenstein)
plt.show()
```

```python
mapache_frankenstein = np.concatenate((mapache_parte_uno,mapache_parte_dos),axis=0)
plt.imshow(mapache_frankenstein)
plt.show()
```
