```python
import numpy as np
np.histogram([1,2,1],bins = [0,1,2,3])
mu, sigma = 2,0.5
data = np.random.normal(mu,sigma,10000)
```

```python
print(data)
```

```python
(n, bin_edges) = np.histogram(data,bins=50)
```

```python
from matplotlib import pyplot
```

```python
pyplot.plot(bin_edges[1:],n)
pyplot.show()
```
